### A function that list all the files inside a directory
And also print print all filenames in subdirectories without using `os.walk`


Write a version of os.walk (4without using os.walk), starting with a directory and going through each of its subdirectories one at a time, printing all of the files it contains.

```python
import glob
# glob.glob can use patterns like * ... and returns full pathname 
from os.path import join , isdir# a version of join that works with filesystem and can work with slash and everything
def print_filename(dirname):
    all_filenames = glob.glob(join(dirname, '*'))

    for one_filename in all_filenames:
        if isdir(one_filename):
            print(f'\t**{one_filename}')
            all_filenames+=(glob.glob(join(one_filename, '*')))
        else:
            print(one_filename)

print_filename("C:\Program Files\Asus")
```
